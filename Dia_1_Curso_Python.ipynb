# Tipos de Datos

```python
# Existen los datos:
enteros = 4 # Son numeros sin decimales
flotantes = 3.9 # Son numeros con decimales
type(flotantes)
```

```python
# Caracteres/ Stings
caracteres = "a" #1 solo
cadena_caracteres ="Hola mundo" #todos los caracteres que quiera agregar o incluir en este espacio
type(caracteres)
```

```python
# Booleanos
# Python tiene palabras reservados que los marca de color azul en este caso
falso = False # falso es = 0
verdadero = True # verdadero es = 1
type(verdadero)
```

# El usuario tambien puede ingresar datos


```python
# Con la funcion input, el usuario puede ingresar caulquier tipo de datos
# se puedes guardar los datos en variables
nombre = input("Ingrese su nombre: ")
print("Hola",nombre)
# Vamos a generar un error

num = input("Ingrese un numero: ")
print(num + 1 ) # Este crea un error porque no puedo sumar un string y un numero
```

```python
num = int(input("Ingresa un numero: ")) #int es tipo de dato que estoy ingresando
print(num + 1 )
```

# Operaciones matematicas

```python
# Usa jerarquia de operaciones
# Suma Ordinaria
suma = 1 + 1
print(suma)
variable1, variable2 = 1, 2
print(variable1 + variable2)

# Resta Ordinaria
resta = 1 - 1
print(resta)
variable1, variable2 = 1, 2
print(variable1 - variable2)

#Tercer nivel de importancia
# Multiplicaion ordinaria
Multi = 1 * 2
print(Multi)
```

# Operaciones matematicas

```python
# Usa jerarquia de operaciones
# Suma Ordinaria
suma = 1 + 1
print(suma)
variable1, variable2 = 1, 2
print(variable1 + variable2)

# Resta Ordinaria
resta = 1 - 1
print(resta)
variable1, variable2 = 1, 2
print(variable1 - variable2)
```

```python
#Tercer nivel de importancia
print("Multiplicaion ordinaria")
Multi = 2 * 3.14
print(Multi)
print("----------------")
print("Division ordinaria")
Div = 1 / 2
print(Div)
print("Division de piso") #Redondea la division hacia abajo de numeros entereos, para casos especiales
Div_piso = 1 // 2
print(Div_piso)
```

```python
#Potencias
print("Potencia ordinaria")
potencia = 2 ** 3 #Tambien podemos hacer las raices por como funciones las raices
print(potencia)
```

```python
#Area de un circulo
pi = 3.14
r = 4
A = pi*r**2
print(A)
```

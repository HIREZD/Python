## 3.- self, __ init __ y el ciclo de vida de un objeto

```python
class Perro:    
# Self es la refgerencio o instancia actial, __ init __
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    def presentarse(self):
        print(f"Soy {self.nombre} y tengo {self.edad} años")
maylah = Perro("Maylah", 5)
maylah.presentarse()
```

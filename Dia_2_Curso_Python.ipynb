## Elif
Ahora hagamos un programa en eque preguntamos la calificaion del 0 al 100 y queremos saber su equivalencia en el sistema de evaluacion americano. Para simplificar, nos basaremos en esto:
A+ = 100
A = 90 - 99
B = 80 - 89
C = 70 - 79
D = 60 - 69
F = < 60


```python
cal = int(input("Ingresa tu calificacion:"))
if cal == 100:
  print("A+")
elif cal >=90:
  print("A")
elif cal >=80:
  print("B")
elif cal >=70:
  print("C")
elif cal >=60:
  print("D")
else:
  print("F")
```

## Ciclo for
Los llamados ciclos for nos ayudan a realizar acciones durante una numera determinada de veces


```python
#imprimr los numero del 0 al 4
for i in range(5): #todos los ciclos empiezan en 0, al igual a una matriz el primer numero es 0
  print(i) # i es una variable temporal
```

```python
for i in range(2,11):
  print(i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

```python
#range con el paso, range inicio, fin, paso
for i in range (0,20,2):
  print(i)
```

## While


```python
# Es muy similar al for, solo que en for tu tienes que decidir el rango, while es un bucle es infinito hasta que la condicion sea falsa y se salga del bucle
suma = 0
numeros = int(input("Ingresa le numero a sumar o 0 para salir:"))

while numeros != 0: #!= significa que es distinto
  suma += numeros
  numeros = int(input("Ingresa le numero a sumar o 0 para salir:"))

print(f"Tu suma es: {suma}")
```

```python
# usar el while como for

numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```

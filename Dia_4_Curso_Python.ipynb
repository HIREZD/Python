## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo  = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida

print(mensaje_completo("Hiram"))
```

```python
# Ahora supongamos que tenemos una heladeria

def pedir_sabor():
  sabores = ["vainilla", "chocolate", "napolitano"]
  print(f"Sabores disponiblese: ",sabores)
  #lowwer() hace minisculas las palabras
  #strip() elimina los espaciones tanto antes como despues
  sabor = input("Eligue un sabor: ").strip().lower()

# Si sabor NO esta en sabores
  while sabor not in sabores:
    sabor = input("Eligue un sabor disponible: ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve quieres: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve quieres: (ingrese un valor valido)"))
  return cantidad

# Funcion Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderas: "))
for _ in range(total_pedidos):
  pedir_helado()



```

## Funciones Lambda
Una funcion lambda es una funcion acortada para escribir una funcion de esta forma:

nombreFuncion = lambda parametros: accion

La diferencia entre las funciones normales es que suelen ser un poco mas limitadas

```python
sumar = lambda x,y : x+y
print(sumar(2,2))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.4, 500)
print("El precio a pagar es:", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio[1])
```

```python
def promediosYmayor(promedios):
    promedio = 0
    for x in promedios:
        promedio += x

    promedio = promedio / len(promedios)
    return promedio, max(promedios)


promedios = []
continuar = ""

while True:
    if len(promedios) > 0:
        continuar = ""
        while continuar not in ("si", "no"):
            continuar = input("¿Desea ingresar otra calificación? (si/no): ").lower()

        if continuar == "no":
            break

    while True:
        try:
            calif = int(input("Indique la calificación: "))
            if calif < 0 or calif > 100:
                print("Ingrese un valor válido (0 a 100)")
            else:
                promedios.append(calif)
                break
        except ValueError:
            print("Ingrese un valor numérico")

prom, mayor = promediosYmayor(promedios)

print("\nResultados:")
print("Promedio:", prom)
print("Mayor:", mayor)
```

```python
# Calculadora -> Suma, resta, producto, division y potencia

def suma(a, b):
  return a + b
def rsta(a, b):
  return a - b
def mult(a, b):
  return a * b
def div (a, b):
  if b == 0:
    print("Error: no se puede dividir entre 0")
  return a / b
def pot(a,b):
  return a ** b

def menu():
  print("------------- MI CALCULADORA -------------")
  print("1.- Suma")
  print("2.- Resta")
  print("3.- Multiplicacion")
  print("4.- Division")
  print("5.- Potencia")
  print("6.- Salir")

while True:
  menu()
  try:
    op = int(input("Ingrese una opcion: "))
    if op < 1 or op > 6:
      print("Error: Ingrese una opcion valida")
      continue
  except ValueError:
    print("Error: Ingrese un numero")
  if op == 6:
    print("Gracias por usar mi calculadora")
    break

  try:
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Error: Ingrese un numero valido")

  if op == 1:
    print("Resultado:", suma(a,b))
  elif op == 2:
    print("Resultado:", rsta(a,b))
  elif op == 3:
    print("Resultado:", mult(a,b))
  elif op == 4:
    print("Resultado:", div(a,b))
  elif op == 5:
    print("Resultado:", pot(a,b))
  else :
    print( "Ingrese una opcion valida")
```

# Estructuras de datos

Las estructuras de datos son formas deorganizar informaci√≥n para trabajar con ellas de manera eficiente

- Listas -> Ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave:valor

Nota: existen mas tipos de datos

- Pilas (FILO -> First in, Last Out)
- Colas (FIFO -> First in, First out)
- Listas enlazadas (o doblemente enlazado)

```python
# len() saca el tam√±o de la lista
for i in range(len(lista_de_compras)): #range no trabaja con elementos, tenemos que saber el tama√±o de la lsita con la funcion len
  print(f"{i+1} - {lista_de_compras[i]}") #i+1 imprimo el valor numerico de la lista y agregandole 1 por la anidacion de python que empieza desde 0, y la otra funcion imprime el elemto de la lista en la posicion i
```

```python
promedios = [55.6, 97.5, 23.6, 97.5, 72.4, 100]
print(promedios)
promedios.sort() # Sirve para ordenar la lista de manera ascendente
print(promedios)

# hay muchas formas de ordenar datos
#esta opcion de .sort es muy pesado y es mejor usar algoritmos especializados en eso

#ordenar una lista de manera descendentes
promedios.sort(reverse = True)
print(promedios)
```

```python
# Declaracion y uso de matrices
calificaciones = [[100,88,93], [99,56,72]]
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f"Calificaion {j}: {cali_j}")
    j += 1
  j = 0
  i += 1
```

```python
# Union

matutino = {20322,22183, 123123, 53264}
vespertino = {20322, 55555, 44444, 123542}

#Union con .union
total_matriculas = matutino.union(vespertino)
print(total_matriculas)

# Union con  |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
# Interseccion
ca = {1234, 121, 646 , 12414, 7613527, 4562138}
al = {1234, 111, 222, 33, 1548}

#Interseccion con .intersection
ca_al = ca.intersection(al)
print(ca_al)

#Interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
solo_calculo = ca.difference(al)
print(solo_calculo)

# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
print("Juego: Adivina el N√∫mero")

numero_secreto = int(input("Jugador 1, escribe el n√∫mero secreto (1-20): "))

intentos = 0
max_intentos = 4

while intentos < max_intentos:
    intento = int(input("Adivina el n√∫mero: "))
    intentos += 1

    if intento == numero_secreto:
        print("¬°Correcto!")
        break
    else:
        print("Incorrecto, intenta de nuevo.")
```

```python
# opcional

import random

opciones = ["piedra", "papel", "tijeras"]

while True:
    while True:
        eleccion = input(
            "\nElige piedra, papel o tijeras (o escribe salir si ya no quieres jugar): "
        ).lower()

        if eleccion not in opciones and eleccion != "salir":
            print("Opci√≥n no v√°lida, intenta de nuevo.")
        else:
            break

    if eleccion == "salir":
        print("¬°Gracias por jugar!")
        break

    aleatoria = random.choice(opciones)
    print("La computadora eligi√≥:", aleatoria)

    # Determinamos el ganador
    if eleccion == aleatoria:
        print("Empate üòê")
    elif (
        (eleccion == "piedra" and aleatoria == "tijeras") or
        (eleccion == "papel" and aleatoria == "piedra") or
        (eleccion == "tijeras" and aleatoria == "papel")
    ):
        print("¬°Ganaste! üòÑ")
    else:
        print("Perdiste üò¢")
```

```python
# Lista de productos
productos = ["manzana", "banana", "leche", "pan"]  # <-- LISTA

# Tupla de precios
precios = (10, 5, 20, 15)  # <-- TUPLA

# Diccionario de inventario: producto -> {"precio": x, "stock": y}
inventario = {
    productos[i]: {"precio": precios[i], "stock": 10}
    for i in range(len(productos))
}  # <-- DICCIONARIO

# Set de productos en oferta
ofertas = {"manzana", "pan"}  # <-- SET

# Mostrar productos
for p in productos:
    msg = "OFERTA!!!" if p in ofertas else "Grandes precios!"
    info = inventario[p]
    print(f"{p.title()} - ${info['precio']} - Stock: {info['stock']} -> {msg}")

# Compra simple
compra = input("\n¬øQu√© producto quieres comprar? ").strip().lower()

if compra in inventario and inventario[compra]["stock"] > 0:
    inventario[compra]["stock"] -= 1
    print(f"Compraste 1 {compra}. Stock restante: {inventario[compra]['stock']}")
else:
    print("Producto no disponible.")
```
